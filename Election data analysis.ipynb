```python
# This is the url link for the poll data in csv form
url = "http://elections.huffingtonpost.com/pollster/2012-general-election-romney-vs-obama.csv"

# Use requests to get the information in text form
source= requests.get(url).text

# Use StringIO to avoid an IO error with pandas
poll_data = StringIO(source)
```

We observe that polls in the years 2009 - mid 2012 were leaning towards Obama. We see that the polls had started to favor Romney.

Also observe the two huge spikes favoring Romney near 10/7/2011. Lets mine that further to see what was going on then.


```python
poll_df.head()
```

```python
poll_df['Difference'].min()

```

```python
row_in = 0
xlimit = []

for d in poll_df['Difference']:
    if d == -0.16:
        xlimit.append(row_in)
        row_in +=1
    else:
        row_in +=1

print (xlimit)

poll_df.plot('Start Date','Difference',figsize=(12,4),marker='o',linestyle='-',color='purple',xlim=(98,123))
```

There you go!!! The slump can be explained because of the fact that the presidential debate on 2nd Oct heavily favored Mitt Romney.

Check out the links - http://www.gallup.com/poll/158048/romney-obama-among-likely-voters.aspx

```python
row_in = 0
xlimit = []

for date in poll_df['Start Date']:
    if date[0:7] == '2012-10':
        xlimit.append(row_in)
        row_in +=1
    else:
        row_in +=1

print min(xlimit)
print max(xlimit)
```

```python
top_donor = donor_df['contb_receipt_amt'].copy()

top_donor.sort_values(inplace=True) 

top_donor


```

```python
candidates = donor_df.cand_nm.unique()

candidates
```

```python
# Dictionary of party affiliation
party_map = {'Bachmann, Michelle': 'Republican',
           'Cain, Herman': 'Republican',
           'Gingrich, Newt': 'Republican',
           'Huntsman, Jon': 'Republican',
           'Johnson, Gary Earl': 'Republican',
           'McCotter, Thaddeus G': 'Republican',
           'Obama, Barack': 'Democrat',
           'Paul, Ron': 'Republican',
           'Pawlenty, Timothy': 'Republican',
           'Perry, Rick': 'Republican',
           "Roemer, Charles E. 'Buddy' III": 'Republican',
           'Romney, Mitt': 'Republican',
           'Santorum, Rick': 'Republican'}

# Now map the party with candidate
donor_df['Party'] = donor_df.cand_nm.map(party_map)
```

```python
donor_df = donor_df[donor_df.contb_receipt_amt >0]
```

```python
occupation_df = donor_df.pivot_table('contb_receipt_amt'
                                    ,index='contbr_occupation',
                                    columns ='Party',
                                    aggfunc='sum')
```

```python
occupation_df.head()
```

```python
occupation_df.shape
```

```python
occupation_df = occupation_df[occupation_df.sum(1) > 1000000]
```

```python
occupation_df.shape
```

```python
occupation_df.plot(kind='bar')
```

```python
occupation_df.plot(kind='barh',figsize=(10,13),cmap='seismic')
```

```python
occupation_df.drop(['INFORMATION REQUESTED PER BEST EFFORTS','INFORMATION REQUESTED'],axis=0,inplace=True)
```

```python
occupation_df.loc['CEO'] = occupation_df.loc['CEO'] + occupation_df.loc['C.E.O.']

occupation_df.drop(['C.E.O.'],inplace=True)
```
